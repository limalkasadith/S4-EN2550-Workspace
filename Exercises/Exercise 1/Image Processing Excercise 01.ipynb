```python
#example 1 for broadcasting
a = np.array([1,2,3]) # 1,3
b = np.array([1,2,3]).reshape(3,1)  #3,1
print(a,b,sep="\n")
a+b #resulting shape = 3,3
```

```python
#example 2 for broadcasting
a = np.array([[1,2,3],[1,2,3],[1,2,3]]) # 3,3
b = np.array([1,2,3])  #3,1
print(a,b,sep="\n-----------------\n")
a+b #resulting shape = 3,3
```

```python
#example 3 for broadcasting
a = np.array([[1,2,3],[1,2,3],[1,2,3]]) # 3,3
b = np.array([1])  #1,1
print(a,b,sep="\n-----------------\n")
a+b #resulting shape = 3,3
```

```python
import matplotlib.pyplot as plt
m, c = 2 , 4
N = 1000000
x = np . linspace (0 , N-1, N) . reshape (N, 1 )
sigma = 10
y = m*x + c + np . random . normal (0 , sigma , (N, 1 ) )
plt.plot(x,y)
```

```python
X = np.append(np.ones((N,1),dtype=x.dtype),x,axis=1)
```

```python
from numpy.linalg import inv
inv(np.dot(X.T,X))@X.T@y
```
